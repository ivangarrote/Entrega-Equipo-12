```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.optimizers import SGD
from keras.callbacks import EarlyStopping
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
from keras.utils import to_categorical
from keras.regularizers import l2
from google.colab import files
from sklearn.metrics import accuracy_score, f1_score, classification_report

# Crear caractersticas de rendimiento para cada equipo
def add_team_performance_features(data):
    data['HomeWinRate'] = data.groupby('HomeTeam')['FTR'].transform(lambda x: (x == 'H').mean())
    data['AwayWinRate'] = data.groupby('AwayTeam')['FTR'].transform(lambda x: (x == 'A').mean())
    return data

# Preprocesamiento de datos
uploaded=files.upload()
data = pd.read_excel('Dataset1.xlsx')
data = add_team_performance_features(data)

# Codificacin de la columna de resultado
lb_make = LabelEncoder()
data['categorical FTR'] = lb_make.fit_transform(data['FTR'])

# Seleccin de caractersticas
columns_name = ['FTHG', 'FTAG', 'HTHG', 'HTAG', 'HS', 'AS', 'HST', 'AST', 'HF', 'AF', 'HC', 'AC', 'HY', 'AY', 'HR', 'AR', 'HomeWinRate', 'AwayWinRate']
X_all = data[columns_name]
y_all = to_categorical(data['categorical FTR'])

# Escalado de caractersticas
scaler = MinMaxScaler()
X_all_scaled = scaler.fit_transform(X_all)

# Divisin de los datos en conjunto de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X_all_scaled, y_all, test_size=0.3, random_state=4)

# Definir el modelo de red neuronal
def build_model():
    model = Sequential([
        Dense(64, activation='relu', input_shape=(X_train.shape[1],), kernel_regularizer=l2(0.01)),
        Dropout(0.3),
        Dense(32, activation='relu', kernel_regularizer=l2(0.01)),
        Dropout(0.3),
        Dense(16, activation='relu', kernel_regularizer=l2(0.01)),
        Dense(3, activation='softmax')
    ])

    model.compile(optimizer=SGD(learning_rate=0.001, momentum=0.9),
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])
    return model

# Entrenamiento del modelo
model = build_model()
early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
history = model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=100, verbose=2, callbacks=[early_stopping])

# Funcin para visualizar los resultados
def plot_training_history(history):
    hist = pd.DataFrame(history.history)
    hist['epoch'] = history.epoch

    plt.figure()
    plt.plot(hist['epoch'], hist['loss'], label='Prdida en entrenamiento')
    plt.plot(hist['epoch'], hist['val_loss'], label='Prdida en validacin')
    plt.xlabel('poca')
    plt.ylabel('Prdida')
    plt.legend()

    plt.figure()
    plt.plot(hist['epoch'], hist['accuracy'], label='Precisin en entrenamiento')
    plt.plot(hist['epoch'], hist['val_accuracy'], label='Precisin en validacin')
    plt.xlabel('poca')
    plt.ylabel('Precisin')
    plt.legend()
    plt.show()

plot_training_history(history)

# Evaluacin en el conjunto de prueba
y_pred_probs = model.predict(X_test)  # Obtener probabilidades predichas
y_pred = np.argmax(y_pred_probs, axis=1)  # Convertir probabilidades a class labels
y_test_labels = np.argmax(y_test, axis=1) #Convertir y_test a class labels

test_accuracy = accuracy_score(y_test_labels, y_pred)
f1_weighted = f1_score(y_test_labels, y_pred, average='weighted')
print('Precisin en el conjunto de prueba:', round(test_accuracy, 3))
print('F1 Score (weighted):', round(f1_weighted, 3))
print(classification_report(y_test_labels, y_pred, target_names=['Away', 'Draw', 'Home']))

# Funcin para hacer predicciones de partidos especficos
def predict_match_result(home_team, away_team, data, model, scaler):
    dato = data[(data['HomeTeam'] == home_team) & (data['AwayTeam'] == away_team)]
    if dato.empty:
        print(f"No hay datos para el partido entre {home_team} y {away_team}.")
        return None

    X_example = dato[columns_name]
    X_example_scaled = scaler.transform(X_example)

    pred = model.predict(X_example_scaled)
    pred_probs = [float(format(x, '.3f')) for x in pred[0]]

    # Grfico de torta
    pie_labels = ['Away', 'Draw', 'Home']
    plt.figure()
    plt.pie(pred_probs, labels=pie_labels, autopct='%1.2f%%', startangle=90)
    plt.axis('equal')
    plt.show()

    predicted_outcome = pie_labels[np.argmax(pred_probs)]
    print('Prediccin del ganador:', predicted_outcome)
    return predicted_outcome
```
